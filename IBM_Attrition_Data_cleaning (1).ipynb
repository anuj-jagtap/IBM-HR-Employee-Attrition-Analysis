Dropping useless columns

```python
data=data.drop(columns=['EmployeeCount','Over18','StandardHours'])
```

```python
data.info()
```

Mapping labels to numeric columns

```python
ordinal_maps = {"Education": {1:"Below College", 2:"College", 3:"Bachelor", 4:"Master", 5:"Doctor"},
"EnvironmentSatisfaction": {1:"Low", 2:"Medium", 3:"High", 4:"Very High"},
"JobInvolvement": {1:"Low", 2:"Medium", 3:"High", 4:"Very High"},
"JobSatisfaction": {1:"Low", 2:"Medium", 3:"High", 4:"Very High"},
"PerformanceRating": {1:"Low", 2:"Good", 3:"Excellent", 4:"Outstanding"},
"RelationshipSatisfaction": {1:"Low", 2:"Medium", 3:"High", 4:"Very High"},
"WorkLifeBalance": {1:"Bad", 2:"Good", 3:"Better", 4:"Best"}
}

for col, mapping in ordinal_maps.items():
    data[f"{col}_Label"]=data[col].map(mapping)
```

```python
data.info()
```

Creating Attrition Flag

```python
data["Attrition_Flag"]=data['Attrition'].map({"Yes":1, "No":0})
```

Bin continuous variables

```python
# Age Group
data["AgeGroup"] = pd.cut(data["Age"], bins=[17, 25, 35, 45, 60], labels=["18-25", "26-35", "36-45", "46-60"])

# Income Band
data["IncomeBand"] = pd.qcut(data["MonthlyIncome"], q=4, labels=["Low", "Medium", "High", "Very High"])

# Tenure Group
data["TenureGroup"] = pd.cut(data["YearsAtCompany"], bins=[-1, 2, 5, 10, 40], labels=["0-2 New", "3-5 Mid", "6-10 Established", "10+ Veteran"])

# Distance Band(Home distance)
data["DistanceBand"] = pd.cut(data["DistanceFromHome"], bins=[0, 10, 20, 30], labels=["Near (<10)", "Medium (10-20)", "Far (20+)"], include_lowest=True)
```

# **EDA**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

**Attrition Rate**

```python
# Overall Attrition Rate
Attrition_rate=data['Attrition_Flag'].mean()*100
print(f"Overall Attrition Rate: {Attrition_rate:.2f}%")
```

```python
data["Attrition"].value_counts().plot(kind='bar', title="Attrition Count")
plt.show()
```

We found that Attrition Rate is low which is 16.13%. But it will impact as we compare overall count.

**Distributions & Counts**

```python
data[["Age", "MonthlyIncome", "YearsAtCompany", "DistanceFromHome"]].hist(bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()

# Categorical spread
for col in ["Department", "JobRole", "Gender", "MaritalStatus"]:
    print(data[col].value_counts(), "\n")
```

**Attrition rate by each grouped column**

```python
def attrition_rate_by(col):
    result = data.groupby(col)["Attrition_Flag"].mean().sort_values(ascending=False) * 100
    result.plot(kind="bar", title=f"Attrition Rate % by {col}", color="indianred")
    plt.ylabel("Attrition Rate %")
    plt.xticks(rotation=45)
    plt.show()
    return result

attrition_rate_by("Department")
attrition_rate_by("JobRole")
attrition_rate_by("AgeGroup")
attrition_rate_by("IncomeBand")
attrition_rate_by("TenureGroup")
attrition_rate_by("DistanceBand")
attrition_rate_by("OverTime")
attrition_rate_by("WorkLifeBalance_Label")
attrition_rate_by("JobSatisfaction_Label")
attrition_rate_by("EnvironmentSatisfaction_Label")
attrition_rate_by("MaritalStatus")
```

**Compensation comparison: stayed vs left**

```python
data.groupby("Attrition")["MonthlyIncome"].mean()

sns.boxplot(data=data, x="Attrition", y="MonthlyIncome")
plt.title("Monthly Income: Stayed vs Left")
plt.show()
```

**Correlation heatmap**

```python
data["OverTime_Flag"] = data["OverTime"].map({"Yes": 1, "No": 0})
```

```python
import numpy as np
numeric_cols = data.select_dtypes(include=[np.number]).columns
corr = data[numeric_cols].corr()["Attrition_Flag"].sort_values(ascending=False)
print(corr)

plt.figure(figsize=(6, 10))
sns.heatmap(corr.to_frame(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation with Attrition")
plt.show()
```
